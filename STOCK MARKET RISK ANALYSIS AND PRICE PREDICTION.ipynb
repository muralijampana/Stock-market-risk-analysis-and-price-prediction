```
 
import yfinance as yf

ticker_symbol = 'AAPL'
stock_data = yf.download(ticker_symbol, start='2023-01-01', end='2023-12-31', interval='1d')
stock_data.head()
```

```
stock_data.describe()
```

```
stock_data.info()
```

```
stock_data.isnull().sum()
```

```
 

plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Adj Close'], color='blue')
plt.title('AAPL Adjusted Closing Price (2023)')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price (USD)')
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Volume'], color='orange')
plt.title('AAPL Trading Volume (2023)')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```
stock_data['30 Day MA'] = stock_data['Adj Close'].rolling(window=30).mean()
stock_data['30 Day Std'] = stock_data['Adj Close'].rolling(window=30).std()

plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Adj Close'], label='Adj Close', color='blue')
plt.plot(stock_data.index, stock_data['30 Day MA'], label='30 Day Moving Average', color='red')
plt.fill_between(stock_data.index, stock_data['30 Day MA'] - stock_data['30 Day Std'], 
                 stock_data['30 Day MA'] + stock_data['30 Day Std'], color='gray', alpha=0.2)
plt.title('AAPL Adjusted Close Price with 30-Day Moving Average and Volatility')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```
stock_data['Daily Return (%)'] = stock_data['Adj Close'].pct_change() * 100
plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Daily Return (%)'], label='Daily Return (%)', color='green')
plt.title('AAPL Daily Returns (2023)')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.grid(True)
plt.show()
```

```
# 1. Calculate 30-day Moving Average and Standard Deviation
stock_data['30 Day MA'] = stock_data['Adj Close'].rolling(window=30).mean()
stock_data['30 Day Std'] = stock_data['Adj Close'].rolling(window=30).std()

# 2. Detect Anomalies (Price deviating more than 2 standard deviations from the moving average)
stock_data['Anomaly'] = np.abs(stock_data['Adj Close'] - stock_data['30 Day MA']) > (2 * stock_data['30 Day Std'])

# 3. Plot the Adjusted Close Price and anomalies
plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Adj Close'], label='Adjusted Close Price', color='blue')
plt.plot(stock_data.index, stock_data['30 Day MA'], label='30 Day Moving Average', color='orange')

# Highlight anomalies
anomalies = stock_data[stock_data['Anomaly'] == True]
plt.scatter(anomalies.index, anomalies['Adj Close'], color='red', label='Anomaly', marker='o', s=50)

plt.title('AAPL Stock Price with Anomalies (2023)')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Display the anomalies
print("Anomalies Detected:\n", anomalies[['Adj Close', '30 Day MA', '30 Day Std']])
```

```
from sklearn.ensemble import IsolationForest

# 1. Train Isolation Forest Model
X = stock_data[['Adj Close', 'Volume']].fillna(0)  # Use adjusted close price and volume for anomaly detection
iso_forest = IsolationForest(contamination=0.01, random_state=42)
stock_data['Anomaly_IF'] = iso_forest.fit_predict(X)

# 2. Plot the stock prices and the anomalies detected by Isolation Forest
plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Adj Close'], label='Adjusted Close Price', color='blue')

# Highlight anomalies
anomalies_if = stock_data[stock_data['Anomaly_IF'] == -1]
plt.scatter(anomalies_if.index, anomalies_if['Adj Close'], color='red', label='Anomaly (IF)', marker='o', s=50)

plt.title('AAPL Stock Price with Isolation Forest Anomalies (2023)')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Display anomalies detected by Isolation Forest
print("Anomalies Detected by Isolation Forest:\n", anomalies_if[['Adj Close', 'Volume']])
```

```
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis

 

# Calculate Daily Returns
stock_data['Daily Return'] = stock_data['Adj Close'].pct_change()

# Remove NaN values
daily_returns = stock_data['Daily Return'].dropna()

### 1. Plot Histogram of Daily Returns ###
plt.figure(figsize=(10,6))
plt.hist(daily_returns, bins=50, alpha=0.6, color='blue', density=True)

# Fit a normal distribution to the data
mean_return = daily_returns.mean()
std_return = daily_returns.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_return, std_return)

# Overlay the normal distribution curve
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution Fit')
plt.title('Distribution of AAPL Daily Returns (2023)')
plt.xlabel('Daily Return')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()

### 2. Descriptive Statistics ###
mean = daily_returns.mean()
std_dev = daily_returns.std()
skewness = skew(daily_returns)
kurt = kurtosis(daily_returns)

print(f"Mean Daily Return: {mean:.4f}")
print(f"Standard Deviation (Volatility): {std_dev:.4f}")
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}")
```
